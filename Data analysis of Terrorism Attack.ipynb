```python
import numpy as np

import csv
file_obj=open('terrorismData2.csv')
data=csv.DictReader(file_obj,skipinitialspace=True)


killed = []
wounded = []
country = []

for row in data:
    killed.append(row['Killed'])
    wounded.append(row['Wounded'])
    country.append(row['Country'])
    
    
    
np_killed=np.array(killed)
np_wounded=np.array(wounded)



np_killed[np_killed=='']='0.0'
np_wounded[np_wounded=='']='0.0'



np_killed = np.array(np_killed,dtype=float)
np_wounded=np.array(np_wounded,dtype=float)
killed_wounded= np_killed+  np_wounded
np_country = np.array(country)



bool_arr=np_country=='Pakistan'
#bool_arr
ans1=np_killed[bool_arr]
ans2=np_wounded[bool_arr]
ans3=killed_wounded[bool_arr]
print(np.sum(ans1))
print(np.sum(ans2))
print(np.sum(ans3))



bool_arr=np_country=='India'
#bool_arr
ans4=np_killed[bool_arr]
ans5=np_wounded[bool_arr]
ans6=killed_wounded[bool_arr]
print(np.sum(ans4))
print(np.sum(ans5))
print(np.sum(ans6))



bool_arr=np_country=='Syria'
#bool_arr
ans7=np_killed[bool_arr]
ans8=np_wounded[bool_arr]
ans9=killed_wounded[bool_arr]
print(np.sum(ans7))
print(np.sum(ans8))
print(np.sum(ans9))



bool_arr=np_country=='Afghanistan'
#bool_arr
ans10=np_killed[bool_arr]
ans11=np_wounded[bool_arr]
ans12=killed_wounded[bool_arr]
print(np.sum(ans10))
print(np.sum(ans11))
print(np.sum(ans12))
```
