```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0,5,0.001)
y= x**3
z=y**2

plt.bar(y,z, color="red")
plt.title('Matplotlib Graph demo')
plt.show()
```

```python
x=np.arange(0 ,5,0.01)
y= x**3
y2= x**2

plt.plot(x,y,'black', linewidth='5' ,label='x^3')
plt.plot(x,y2,color='red', linewidth='5', label = 'x^2')

plt.ylabel('y')
plt.xlabel('x')


plt.text(2,80,'text', fontsize=10)
plt.grid()
plt.legend()
plt.title('Matplotlib Graph2')
plt.show()
```
